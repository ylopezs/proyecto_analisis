# LIMPIEZA DE DATOS

````python
# Importar librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Lectura
from google.colab import drive
#drive.mount('/gdrive')

ruta = "https://raw.githubusercontent.com/ylopezs/proyecto_analisis/refs/heads/main/Apropiaci_n_TIC_del_Valle_del_Cauca_20250625.csv"
data = pd.read_csv(ruta)

print(data.head())
print(data.info())
````

````python
data.dropna(inplace=True)
print(data.info())
````

````python
columnas_relevantes = ['Municipio','Sexo','Clasificacion Academica','Entidad Organización','Programa','Modalidad','Beneficio', 'Latitud', 'Longitud']
for col in columnas_relevantes:
  print(f'Columna {col}: {data[col].nunique()} subniveles')
````

````python
fig, ax= plt.subplots(nrows=6, ncols=1, figsize=(10,30))
fig.subplots_adjust(hspace=1)

for i, col in enumerate(columnas_relevantes):
  sns.countplot(x=col, data=data, ax=ax[i])
  ax[i].set_title(col)
  ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation = 90)
````

````python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize = (5,5))
fig.subplots_adjust(hspace=1)
sns.countplot(x='Modalidad', data=data)
plt.xlabel("Modalidad")
plt.ylabel("Cantidad personas")
plt.title("Modalidad de estudio preferida por los beneficiarios")
plt.show()
plt.show()
````

**¿Qué hicimos?**
En este apartado realizamos una aproximación inicial al análisis de datos.
- En un primer momento descubrimos si faltan algunos datos o si hay casillas vacías.
- Luego seleccionamos aquellas columnas que tienen los datos completos.

````python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np

# Crear un mapa del mundo
plt.figure(figsize = (16,12))

# Incializar Basemap con la proyección 'mill' (Miller cylindrical projection)
# Approximate coordinates for Valle del Cauca, Colombia
m = Basemap(projection = 'mill', llcrnrlat = 3.0, urcrnrlat = 5.0, llcrnrlon = -77.0, urcrnrlon = -75.0, resolution = 'f') # Approximate coordinates for Valle del Cauca

# Rellenar los continentes
m.fillcontinents(color= 'lightgreen', lake_color='lightblue')

# Rellenar los océanos
m.drawmapboundary(fill_color = 'lightblue')

# Dibujar costas y países
m.drawcoastlines()
m.drawcountries()
m.drawcounties(linewidth=1.5, color='darkblue') # Add municipal boundaries with increased linewidth and contrasting color
m.drawstates()

# Convertir latitudes y longitudes en coordenadas x, y del mapa
x, y = m(data['Longitud'], data['Latitud'])

# Dibujar los puntos in the map
m.scatter(x, y, marker = 'o', color = 'red', zorder = 5)

# Labels for the points
for lat, lon, city in zip(data['Latitud'], data['Longitud'], data['Sexo']):
  plt.text(m(lon, lat)[0], m(lon,lat)[1], city, fontsize= 12, ha= 'right', color = 'blue')

# Add labels for municipalities
for municipio in data['Municipio'].unique():
    # Get the average coordinates for each municipality
    avg_lat = data[data['Municipio'] == municipio]['Latitud'].mean()
    avg_lon = data[data['Municipio'] == municipio]['Longitud'].mean()
    x_mun, y_mun = m(avg_lon, avg_lat)
    plt.text(x_mun, y_mun, municipio, fontsize=10, ha='center', va='center', color='black')


# Título del gráfico
plt.title("Mapa con Coordenadas Geográficas del Valle del Cauca")
plt.show()
````

````python
# prompt: hazme un grafico de linea para entidad versus organizacion

# Calcular la cantidad de registros por entidad
entidad_counts = data['Entidad Organización'].value_counts()

# Crear el gráfico de líneas
plt.figure(figsize=(12, 6))  # Ajustar el tamaño del gráfico
plt.plot(entidad_counts.index, entidad_counts.values, marker='o', linestyle='-')

plt.xlabel("Entidad Organización")
plt.ylabel("Cantidad de Registros")
plt.title("Cantidad de Registros por Entidad Organización")
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x para mejor legibilidad
plt.grid(True) # Agregar una cuadrícula
plt.tight_layout() # Ajustar el diseño para evitar solapamientos
plt.show()
````

````python
# prompt: hazme graficas de correlacion en  las variables que aplique

# Seleccionar solo las columnas numéricas para el cálculo de la correlación
df_numeric = data.select_dtypes(include=[np.number])

# Calcular la matriz de correlación
correlation_matrix = df_numeric.corr()

# Crear un mapa de calor de la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Matriz de Correlación de Variables Numéricas")
plt.show()
````

````python
# prompt: hazme un grafico 3d para las variables sexo, municipio y beneficio

from mpl_toolkits.mplot3d import Axes3D

# Asegurarse de que las columnas 'Sexo', 'Municipio', y 'Beneficio' son categóricas
# Convertir las columnas categóricas a códigos numéricos para el gráfico 3D
data['Sexo_code'] = data['Sexo'].astype('category').cat.codes
data['Municipio_code'] = data['Municipio'].astype('category').cat.codes
data['Beneficio_code'] = data['Beneficio'].astype('category').cat.codes

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Graficar los puntos
# Puedes usar un pequeño valor aleatorio para dispersar puntos idénticos y hacerlos visibles
ax.scatter(data['Sexo_code'] + np.random.rand(len(data)) * 0.2 - 0.1,
           data['Municipio_code'] + np.random.rand(len(data)) * 0.2 - 0.1,
           data['Beneficio_code'] + np.random.rand(len(data)) * 0.2 - 0.1,
           c=data['Sexo_code'], cmap='viridis', marker='o') # Usar el código de Sexo para el color

# Configurar las etiquetas de los ejes con los nombres originales
ax.set_xlabel('Sexo')
ax.set_ylabel('Municipio')
ax.set_zlabel('Beneficio')

# Configurar los ticks para mostrar los nombres originales
ax.set_xticks(data['Sexo_code'].unique())
ax.set_xticklabels(data['Sexo'].unique(), rotation=45, ha='right')

ax.set_yticks(data['Municipio_code'].unique())
ax.set_yticklabels(data['Municipio'].unique(), rotation=-45, ha='left') # Ajustar rotación y alineación para y

ax.set_zticks(data['Beneficio_code'].unique())
ax.set_zticklabels(data['Beneficio'].unique())

plt.title('Gráfico 3D de Sexo, Municipio y Beneficio')
plt.tight_layout()
plt.show()
````
